Q1.Scrape data for “Data Analyst” Job position in “Bangalore” location.

```python
#importing libraries
from selenium import webdriver
import pandas as pd
import warnings   #importing warning modules for ignoring the warnings
warnings.filterwarnings('ignore')
```

```python
#connecting to webdriver
driver=webdriver.Chrome(r'C:\Program Files (x86)\chromedriver.exe')
```

```python
#opening the url
driver.get('https://www.naukri.com/')
```

```python
#sending keys to job search and location
search_job=driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys('Data Analyst')
search_loc=driver.find_element_by_id('qsb-location-sugg')
search_loc.send_keys('Bangalore')
#clicking on search bar
search_btn=driver.find_element_by_xpath("//button[@class='btn']")
search_btn.click()
```

```python
#closing the chat box,if chatbox appeared
chat=driver.find_element_by_xpath("//div[@class='crossIcon chatBot chatBot-ic-cross']").click()
```

```python
#making empty list for storing to scarping the datas
job_title=[]
job_location=[]
company_name=[]
experience_required=[]

#html tags where have the job titles
titles_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")

#text of job_title inside the above html tags
for t in titles_tags:
    job_title.append(t.text)
    
#html tags where have location
loc_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span")
    
#text of location where above html tags have
for l in loc_tags:
    job_location.append(l.text)
    
#html tags of company name
company_tags=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")

#text of company name which in above html tags
for c in company_tags:
    company_name.append(c.text)
    
#html tags for expereience tags
exp_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']/span[1]")

#text of above html tags
for e in exp_tags:
    experience_required.append(e.text)  
```

```python
#putting scraped data into dataframe
jobs=pd.DataFrame({})
jobs['job_title']=job_title[0:10]
jobs['job_location']=job_location[0:10]
jobs['company_name']=company_name[0:10]
jobs['experience_required']=experience_required[0:10]
jobs
```

```python
#closing driver
driver.close()
```

Q4.Scrape data of first 100 sunglasses listings on flipkart.com.

```python
#connecting to driver
driver=webdriver.Chrome(r'C:\Program Files (x86)\chromedriver.exe')
```

```python
driver.get('https://www.flipkart.com/')
```

```python
#closing the login page
button=driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
button.click()
```

```python
#searching the product name in search bar
search_pro=driver.find_element_by_xpath("//div[@class='_3OO5Xc']/input")
search_pro.send_keys("sunglasses")

#clicking search button in the search button
search_btn=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_btn.click()

#storing class names in variables
brand_class='_2WkVRV'
desc_class='IRpwTa'
price_class='_30jeq3'

#creating empty list store the data
brand=[]
description=[]
price=[]
```

```python
#to  scrape product details in multiple pages using for loops and scraping each pages product details
for page in range(1,4,1):
    #page link
    page_url="https://www.flipkart.com/search?q=sunglasses&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=" + str(page)
    #connecting to each page
    driver.get(page_url) 
    #brand name of sunglasses
    title=driver.find_elements_by_class_name(brand_class)
    
    #description of each sunglasses
    specs=driver.find_elements_by_class_name(desc_class)
    
    #prices of each sunglasses
    prices=driver.find_elements_by_class_name(price_class)
    
    #text of brand name
    for b in title:
        brand.append(b.text)
        
    #text of description of sunglasses
    for s in specs:
        description.append(s.text)
        
    #text for prices of sunglasses
    for p in prices:
        price.append(p.text)    
```

```python
#creating dataframe for details of  100 sungalsses
sun_ds=pd.DataFrame({})
sun_ds["Brand"]=brand[0:100]
sun_ds["Description"]=description[0:100]
sun_ds["Price"]=price[0:100]
sun_ds
```

```python
#closing the driver
driver.close()
```

Q6.Scrape data for first 100 sneakers from flipkart.com.

```python
#connecting to driver
driver=webdriver.Chrome(r'C:\Program Files (x86)\chromedriver.exe')
```

```python
#opening the url
driver.get('https://www.flipkart.com/')
```

```python
#closing the login page which appears while opening
button=driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
button.click()
```

```python
#searching sneakers on search fiels
search_box=driver.find_element_by_xpath("//div[@class='_3OO5Xc']/input")
search_box.send_keys("sneakers")
#clicking on search button
search_button=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_button.click()
```

```python
#creating empty list
brand_s=[]
description=[]
descri=[]
Price=[]
Discount=[]
#connecting to each pages
for next_ in range(1,4):
    #page url
    page_links="https://www.flipkart.com/search?q=sneakers&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page="+str(next_)
    
    #connecting to each page
    driver.get(page_links)
    #brand of sneakers
    brand=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
    #description of sneakers
    descr=driver.find_elements_by_xpath("//a[@class='IRpwTa']")#shows only some description
    descrs=driver.find_elements_by_xpath("//a[@class='IRpwTa _2-ICcC']")#shows other description
    #price of the sneakers
    mrp=driver.find_elements_by_xpath("//div[@class='_30jeq3']")
    #discount of sneakers
    off=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']//span")
    
    #text for brand
    for bs in brand:
        brand_s.append(bs.text)
    #text for price
    for pr in mrp:
        Price.append(pr.text)
    #text for discount
    for of in off:
        Discount.append(of.text)
    #text for  descriptions
    for des in descr:
        description.append(des.text)
    for de in descrs:
        descri.append(de.text)
#storing all description in one variable using extend function
description.extend(descri)
```

```python
#checking length of lists
print(len(description),len(brand_s),len(Price),len(Discount))
```

```python
#storing details of 100 sneakers in dataframe
sneakers=pd.DataFrame({})
sneakers['Brand']=brand_s[0:100]
sneakers['Product_Description']=description[0:100]
sneakers['Price']=Price[0:100]
sneakers['Discount']=Discount[0:100]
sneakers
```

```python
#closing the driver
driver.close()
```

```python
#creating empty lists
company_s=[]
short_desc=[]
PRICE=[]
#connecting to multiple pages
for near in range(1,3):
    ne=pagination+str(near)
    #opening pages url
    driver.get(ne)
    #html tags
    #brand of shoes
    s_brand=driver.find_elements_by_class_name("product-brand")
    #description of shoes
    short=driver.find_elements_by_xpath("//h4[@class='product-product']")
    #price of the shoe
    ruppee=driver.find_elements_by_class_name("product-price")
    
    #texts for above tags
    #brand
    for br in s_brand:
        company_s.append(br.text)
    #description
    for sh in short:
        short_desc.append(sh.text)
    #price
    for ru in ruppee:
        PRICE.append(ru.text)
```

```python
print(len(company_s),len(short_desc),len(PRICE))
```

Q10.Scrape the salary data for Data Scientist designation.

```python
#connecting to driver
driver=webdriver.Chrome(r'C:\Program Files (x86)\chromedriver.exe')
```

```python
#opening url in webdriver
driver.get("https://www.ambitionbox.com/")
```

```python
#clicking on salaries option
sal_opt=driver.find_element_by_xpath("//a[@class='link salaries']").click()
#searching job profile 
job_pro=driver.find_element_by_id('jobProfileSearchbox')
job_pro.send_keys('Data Scientist')
```

There is no search button in the search bar so I am just clicking on it manually.

```python
#creating empty list
company_name=[]
salary_record=[]
average=[]
minimum_salary=[]
maximum_salary=[]
experience=[]

#html tags
#company_name
ins=driver.find_elements_by_xpath("//div[@class='name']/a")
#salary record
sal=driver.find_elements_by_xpath("//div[@class='name']/span")
#avearge salary
avg=driver.find_elements_by_xpath("//p[@class='averageCtc']")
#minimum salary
mi=driver.find_elements_by_xpath("//div[@class='value body-medium']")
#maximum salary
#slicing the list of minimum salary we got the maximum salaries
#tags for experience 
exp=driver.find_elements_by_xpath("//div[@class='salaries sbold-list-header']")

#text for above tags
#company name
for i in ins:
    company_name.append(i.text)
#salary record
for s in sal:
    salary_record.append(s.text)
#avearge salary
for a in avg:
    average.append(a.text)
#minimum salary
for m in mi:
    minimum_salary.append(m.text)
#text for experience
for e in exp:
    experience.append(e.text.replace('Data Scientist','').lstrip().replace('\n','').replace('n2','').replace('n3','').replace('\n4',''))
```

```python
#maximum salary
maximum_salary=minimum_salary[1:20:2]
```

```python
#creating dataframe to store data of 10 companies
dataframe=pd.DataFrame({})
dataframe['company_name']=company_name
dataframe['total_salary_record']=salary_record
dataframe['average_salary']=average
dataframe['minimum_salary']=minimum_salary[0:20:2]
dataframe['maximum_salary']=maximum_salary
dataframe['experience_required']=experience
dataframe
```
