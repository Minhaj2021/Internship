```python
#connecting to webdriver
driver1=webdriver.Chrome(r'C:\Program Files (x86)\chromedriver.exe')
#opening the url
driver1.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
time.sleep(5)
driver1.maximize_window()
```

```python
#rank of vedio
rank_vedio=[]
try:
    rank_y=driver1.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[1]")
    for y in rank_y:
        rank_vedio.append(y.text)
except:
    rank_vedio.append('-')
    
#name
vedio_name=[]
try:
    vedio_y=driver1.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[2]")
    for vn in vedio_y:
        vedio_name.append(vn.text)
except:
    vedio_name.append('-')

#artist
artist=[]
try:
    artist_name=driver1.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[3]")
    for art in artist_name:
        artist.append(art.text)
except:
    artist.append('-')
    
#upload date
upload_date=[]
try:
    upload=driver1.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[5]")
    for up in upload:
        upload_date.append(up.text)
except:
    upload_date.append('-')

#views
vedio_views=[]
try:
    views_v=driver1.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[4]")
    for vi in views_v:
        vedio_views.append(vi.text)
except:
    vedio_views.append('-')
```

```python
print(len(rank_vedio),len(vedio_name),len(artist),len(upload_date),len(vedio_views))
```

```python
#view all datasets
view_dataset=driver10.find_element_by_xpath("//span[@class='whitetext']/a/font/b")
time.sleep(2)
view_dataset.click()
```

```python
#datset name
dataset_name=[]
try:
    dataset=driver10.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]")
    for data in dataset:
        dataset_name.append(data.text)
except:
    dataset_name.append('-')

#data type
data_type=[]
try:
    datatype_=driver10.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]")
    for d in datatype_:
        data_type.append(d.text)
except:
    data_type.append('-')

#default task
default_task=[]
try:
    default=driver10.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]")
    for de in default:
        default_task.append(de.text)
except:
    default_task.append('-')

#attribute types
attribute_type=[]
try:
    attribute=driver10.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]")
    for at in attribute:
        attribute_type.append(at.text)
except:
    attribute_type.append('-')

#no of instances
instances_num=[]
try:
    instance=driver10.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]")
    for i in instance:
        instances_num.append(i.text)
except:
    instances_num.append('-')

#no of attributes
attributes_num=[]
try:
    num_a=driver10.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]")
    for na in num_a:
        attributes_num.append(na.text)
except:
    attributes_num.append('-')

#year
year_dataset=[]
try:
    year_d=driver10.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]")
    for y in year_d:
        year_dataset.append(y.text)
except:
    year_dataset.append('-')
```

```python
uci=pd.DataFrame({})

uci['Dataset name']=dataset_name[1:623]
uci['Datatype']=data_type[1:623]
uci['Default Task']=default_task[1:623]
uci['Attribute Type']=attribute_type[1:623]
uci['No of Instances']=instances_num[1:623]
uci['No of Attributes']=attributes_num[1:623]
uci['Year']=year_dataset[1:623]

uci.head(15)
```
